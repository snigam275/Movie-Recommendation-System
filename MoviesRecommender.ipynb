```python
# Now from the above dataframe which we created just now we will create another dataframe which will include 
# columns like movies id, title and tags [ tags will create by merging of columns like overviews,genres,keywords,cast,crew ]
#Data preprocessing
```

```python
movie_data.isnull().sum()
```

```python
movie_data.dropna(inplace = True)
```

```python
movie_data.duplicated().sum()
```

```python
#here we can see that columns like genres keywords cast crew overview all are not in correct format to make a tag because  
# for concatenate we need a list but here they are given in form of list of dictinories so we will create a conversion function
#But here one probelm we will face that these all columns contains data in form of string where we can not directly perform conversion so we 
# to use them we need to convert them as list and that's why we will use function "ast.literal_eval()"
import ast
def convert(obj):
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    
    l = []
    for i in obj:
        if isinstance(i, dict):
            l.append(i['name'])
        else:
            l.append(i)
    return l
```

```python
movie_data['genres'] = movie_data['genres'].apply(convert)
```

```python
def convert1(obj1):
    L = []
    for i in ast.literal_eval(obj1):
        L.append(i['name'])
    return L
```

```python
movie_data['keywords'] = movie_data['keywords'].apply(convert1)
```

```python
# Now for the cast column but we only want top 3 cast so we will make another function and use a counter there

def convert2(obj2):
    Li = []
    counter = 0
    for i in ast.literal_eval(obj2):
        if counter != 4:
            Li.append(i['name'])
            counter += 1
        else:
            break
    return Li
```

```python
movie_data['cast'] = movie_data['cast'].apply(convert2)
```

```python
#Now here for crew and for this we only want that dictonary to fetch name where jod value is director 
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie_data['crew'] = movie_data['crew'].apply(fetch_director)
```

```python
# We need to convert the overview column also 
movie_data['overview'] = movie_data['overview'].apply(lambda x:x.split())
```

```python
movie_data
```

```python
# Now only task is left "Concatination and comverting them into list / means tag " so we will apply transformation
#because here for eg sam worthington and sam mendes both are different but when we list them and model will train itself 
# it will get confuse that which sam because, sam will act as object of the list only so model will confuse which sam 
# that's why we will remove the spaces and make a name for one person like this sam worthington -> samworthington

movie_data['genres'] = movie_data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data['keywords'] = movie_data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data['cast'] = movie_data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data['crew'] = movie_data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data
```

```python
#Final step for tag
movie_data['tag'] = movie_data['overview'] + movie_data['genres'] + movie_data['keywords'] + movie_data['cast'] + movie_data['crew']
```

```python
movie_data
```

```python
new_df = movie_data[["movie_id","title","tag"]]
new_df
```

```python
#now convert tag from list to string
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag'][2]
```

```python
# Convert everything into lower case as it recommendate
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
# text-vectorisation using bag of full words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english') 
```

```python
vector = cv.fit_transform(new_df['tag']).toarray()
vector
```

```python
cv.get_feature_names_out()
```

```python
#Here we will calculate cosine ditance between them rather than using euclidean distance because it is not realiable in higher distance
# We will use cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
#Simalirity score
similarity[0]
```

```python
#Function for recommendation
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:7]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Batman Begins')
```
